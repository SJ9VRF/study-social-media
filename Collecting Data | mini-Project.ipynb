```python
import os
import tweepy as tw
import pandas as pd
import json
from datetime import datetime
```

```python
root_path = '/Users/arefeh/Desktop/tweet_api_new/'
```

```python
consumer_key= '**********'
consumer_secret= '**********'
access_token= '**********'
access_token_secret= '**********'
```

```python
auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
date = list(pd.date_range(start="2020-01-10",end="2021-01-10"))
date_list = [str(item).split()[0] for item in date]
```

```python
search_words = ["#COVIDVaccination", "#antiVax", "anti-vaccine", "#Covid AND #Fake","#Covid","#Corona","#Covid19","#Covid-19"]
```

```python
# Search Key words
search_word = '#CNN AND #Covid'
start_index = 0
for ftem in ['2020-04-08','2020-02-24']:
    print('start_time: '+str(start_index))
    start_index = start_index + 1
    try:
        # Collect tweets
        date_since = ftem
        tweets = tw.Cursor(api.search,
                      q=search_word,
                      lang="en",
                      since=date_since).items(10000)
        print(ftem)
        main_path = root_path +str(search_word)+'/'+ftem+'/'
        n = -1
        for tweet in tweets:
            tweet_item = {}
            tweet_item['id'] = str(tweet.id)
            tweet_item['text'] = tweet.text
            tweet_item['created_date'] = str(tweet.created_at)
            tweet_item['screen_name'] = tweet.user.screen_name
            tweet_item['location'] = tweet.user.location
            tweet_item['followers_count'] = tweet.user.followers_count
            tweet_item['following'] = tweet.user.following
            tweet_item['followers'] = str(tweet.user.followers)

            n = n + 1
            path_dictionary = main_path + str(n)+'.txt'

            json.dump(tweet_item, open(path_dictionary,'w'))


            if (n%10) == 0:
                print(n)
    except Exception:
        pass
```
